```python
#sorting in ascending order
a = [ 3, 5, 67, 58 , 34 , 67, 89, 0 , 1 , 2 , 35 , 34, 99 , 100 , 25  ]
a.sort()
print(a)

```

```python
# sorting in desending order
a = [ 3, 5, 67, 58 , 34 , 67, 89, 0 , 1 , 2 , 35 , 34, 99 , 100 , 25  ]
a.sort()
a.reverse()
print(a)
```

```python
a = [ 3, 5, 67, 58 , 34 , 67, 89, 0 , 1 , 2 , 35 , 34, 99 , 100 , 25  ]
a = sorted(a, reverse=True)
print(a)
```

```python
# nested list 
a=[1,2,3]
print(a)
a.sort(reverse=True)
new_list = [4,5,6 ]
new_list.sort(reverse=True)
print(new_list)
new_list.append(7)
nested_list = [a, new_list]
print(nested_list)
```
